```python
import numpy as np
import matplotlib.pyplot as plt
import scipy

dt = 1/60
fs = 1/dt

w0 = 6
t0 = 2*np.pi/w0

x_upper = 2**8
x_lower = -x_upper
```

```python
vline_names = ['f = 3/π, w0 = 6', 'f = 1/24 Hours', 'f = 1/7 Days']
wavelet_period_names = ['T = π/3', 'T = 24 Hours', 'T = 7 Days', ]

t = np.array([t0, 24, 24*7])
scales = t/t0

FONTSIZE = 32
```

```python
x = np.linspace(start = x_lower, 
                   stop = x_upper, 
                   num = int(2*x_upper*fs))
```

```python
fig, axes = plt.subplots(len(scales), 2, figsize=(30, 10*len(scales)))
for n, scale in enumerate(scales):
    
    w = 2*np.pi / (scale * t0)
    cos_scaled = np.cos(w * x)
    sin_scaled = np.sin(w * x)
    gau_scaled = np.exp(-(w * x / w0)**2 / 2)
    
    sig = (cos_scaled + 1j * sin_scaled) * gau_scaled
    axes[n, 0].plot(x, sig.real, x, sig.imag)

    axes[n, 0].set_xlim([x_lower/2**(len(scales)-n-1), x_upper/2**(len(scales)-n-1)])
    axes[n, 0].set_ylim([-1, 1])
    axes[n, 0].text(x_upper/2**(len(scales)-n-1)*0.5, 0.9, wavelet_period_names[n], fontsize=FONTSIZE)
    axes[n, 0].legend(['Real', 'Imaginary'], loc='upper left', fontsize=FONTSIZE)
    axes[n, 0].tick_params(axis='x', labelsize=20)
    axes[n, 0].tick_params(axis='y', labelsize=20)
    
    filt_fft = np.fft.rfft(sig)
    freq = np.fft.rfftfreq(sig.size, d = dt)
    filt_fft /= np.abs(filt_fft).max()
    
    axes[n, 1].plot(freq, np.abs(filt_fft)**2)
    axes[n, 1].set_xlim([0, 2/t[n]])
    axes[n, 1].set_ylim([0, 1])
    axes[n, 1].vlines(x = 1/t[n], ymin = 0, ymax = 1, colors='red', linestyles='--')
    axes[n, 1].tick_params(axis='x', labelsize=20)
    axes[n, 1].tick_params(axis='y', labelsize=20)
    axes[n, 1].legend(['Power', vline_names[n]], loc='upper left', fontsize=FONTSIZE)
    axes[n, 1].grid(True)

axes[len(scales) - 1, 0].set_xlabel('Time (samples)', fontsize=FONTSIZE)
axes[len(scales) - 1, 1].set_xlabel('Frequency (radians)', fontsize=FONTSIZE)
axes[0, 0].set_title('Wavelet', fontsize=FONTSIZE)
axes[0, 1].set_title(r'|FFT(Wavelet)|$^2$', fontsize=FONTSIZE)

fig.tight_layout()
plt.savefig('wavelet_examples.png', dpi=300)
```
